Mathematical Representation of Model Evaluation:
$$
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

$$
MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$

$$
R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
$$
where:
 $$
 y_i = \text{actual tumor area}
 $$

$$
\hat{y}_i = \text{predicted tumor area }
$$

$$
 SS_{res}= \text{sum of squared residuals}
$$

$$
 SS_{tot} = \text{total sum of squares}
$$

```python
# import libraries and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Load and Inspect the Data
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
columns = ['ID', 'Diagnosis', 'Radius Mean', 'Texture Mean', 'Perimeter Mean', 'Area Mean', 'Smoothness Mean',
           'Compactness Mean', 'Concavity Mean', 'Concave Points Mean', 'Symmetry Mean', 'Fractal Dimension Mean',
           'Radius SE', 'Texture SE', 'Perimeter SE', 'Area SE', 'Smoothness SE', 'Compactness SE',
           'Concavity SE', 'Concave Points SE', 'Symmetry SE', 'Fractal Dimension SE',
           'Radius Worst', 'Texture Worst', 'Perimeter Worst', 'Area Worst', 'Smoothness Worst',
           'Compactness Worst', 'Concavity Worst', 'Concave Points Worst', 'Symmetry Worst', 'Fractal Dimension Worst']

data = pd.read_csv(url, header=None, names=columns)
```

Based on inspection, we notice that there are no missing variables. 

```python
# Preprocess the Data
# Drop ID column (not needed for modeling)
data.drop(columns=['ID'], inplace=True)

# Convert Diagnosis to binary (Malignant=1, Benign=0)
data['Diagnosis'] = data['Diagnosis'].map({'M': 1, 'B': 0})

```

```python
# Define target and predictors
target = 'Area Mean'
predictors = [col for col in data.columns if col not in ['Area Mean', 'Diagnosis']]
```

```python
# Split dataset
X_train, X_test, y_train, y_test = train_test_split(data[predictors], data[target], test_size=0.2, random_state=42)

# Train linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Model evaluation
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"R-squared: {r2}")

```

# ðŸ“ŒTask 4: Apply Linear Regression

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split

# Define target and predictor variables
target = 'Area Mean'
predictors = [col for col in data.columns if col not in ['Area Mean', 'Diagnosis']]

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data[predictors], data[target], test_size=0.2, random_state=42)

# Add a constant term for the intercept in the regression model
X_train_const = sm.add_constant(X_train) 

# Fit the model using statsmodels OLS regression
model_sm = sm.OLS(y_train, X_train_const).fit()

# Display model summary
print(model_sm.summary())
```

When we first performed linear regression, we used all of the features, now we  will add them one at a time using comments to remove any predictors with p values > 0.05, to determine which combination of predictors is linearly related to our target __Area Mean__.

```python
# Define target and predictor variables
target = 'Area Mean'
selected_predictors = ['Radius Mean'
                       ,
                       'Perimeter Mean',
                       'Compactness Mean',
                       #'Concavity Mean',
                       #'Fractal Dimension Mean',
                       #'Radius SE',
                       'Texture SE',
                       'Perimeter SE',
                       'Area SE',
                       'Smoothness SE',
                       #'Compactness SE',
                       'Concavity SE',
                       'Concave Points SE',
                       #'Symmetry SE', 
                      # 'Fractal Dimension SE',
                       'Radius Worst',
                       'Texture Worst',
                       'Perimeter Worst', 
                       'Area Worst',
                       #'Smoothness Worst',
                       #'Compactness Worst',
                       #'Concavity Worst',
                       'Concave Points Worst',
                       'Symmetry Worst',
                       'Fractal Dimension Worst'
                      ]  # Adding Concave Points Mean

# Split dataset using selected predictors
X_train_selected, X_test_selected, y_train, y_test = train_test_split(
    data[selected_predictors], data[target], test_size=0.2, random_state=42
)

# Add a constant term for the intercept in the regression model
X_train_const_selected = sm.add_constant(X_train_selected)

# Fit the model using statsmodels OLS regression
model_sm_selected = sm.OLS(y_train, X_train_const_selected).fit()

# Display model summary
print(model_sm_selected.summary())
```

# Task 5: Performance Evaluation Metrics

```python
# Predict on the test set
X_test_const_selected = sm.add_constant(X_test_selected)
y_pred = model_sm_selected.predict(X_test_const_selected)

# Calculate Mean Squared Error (MSE) and Mean Absolute Error (MAE)
mse = np.mean((y_test - y_pred) ** 2)
mae = np.mean(np.abs(y_test - y_pred))

# Calculate R-squared (coefficient of determination)
r2 = model_sm_selected.rsquared

# Display results
print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (RÂ²): {r2}")
```

# Task 6: Residual Analysis

```python
# Residual analysis
residuals = y_test - y_pred
plt.figure(figsize=(10, 6))
sns.histplot(residuals, bins=30, kde=True)
plt.title("Residual Distribution")
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.title("Residual Plot")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.show()

# Report results
report = {
        "Regression Performance": {
        "MSE": mse,
        "MAE": mae,
        "R2": r2
    }
}
print(report)
```
